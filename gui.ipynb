- create an instance of the environment
- add elements that will be contained in it
- end the loop ➰ 

```python
#creating the environment 
root = tk.Tk()

#defining properties/attribute of the environment
root.geometry("400x500")
root.title("virtual assistant")

#filling the environment with text,pictures or icon

image=Image.open("ladyface.png")
image=image.resize((450,300))
img = ImageTk.PhotoImage(image)



my_label=tk.Label(image=img)
my_label.pack()

my_text = tk.Label(root,text="Lady is active",font=("Helvetica",25),fg="black",bg="#3d6466")
my_text.pack(pady=20)

#closing the environment to allow it run
root.mainloop()
```
